```python
from __future__ import unicode_literals
import youtube_dl


ydl_opts = {
    'format': 'bestaudio/best',
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'wav',
        'preferredquality': '192',
    }],
}
def download_audio(link):
    with youtube_dl.YoutubeDL(ydl_opts) as ydl:
        ydl.download([link])
download_audio('https://www.youtube.com/watch?v=ruAi4VBoBSM')
```

```python
import urllib
from bs4 import BeautifulSoup
searchTags = ['tiesto','armin van buuren','afrojack','steve aoki','fatboy slim','hardwell','aviici','chainsmokers',
             'david guetta','marshmello','don diablo','kshmr','oliver heldens','w&w','calvin harris','skrillex',
             'alan walker','r3hab','alok','dash berlin','dimitri vegas & like mike','deadmau5','zedd']
newSearchTags = []
for searchTag in searchTags:
    newSearchTags.append(searchTag + " audio")
    
urls = []
def search_youtube(text):
    query = urllib.parse.quote(text)
    url = "https://www.youtube.com/results?search_query=" + query
    response = urllib.request.urlopen(url)
    html = response.read()
    soup = BeautifulSoup(html)
    for vid in soup.findAll(attrs={'class':'yt-uix-tile-link'}):
        if not vid['href'].startswith("https://googleads.g.doubleclick.net/"):
            address = 'https://www.youtube.com' + vid['href']
            urls.append(address)
            print('https://www.youtube.com' + vid['href'])

for tag in newSearchTags:
    search_youtube(tag)
    
print(urls)
```

```python
for url in urls:
    download_audio(url)
```
